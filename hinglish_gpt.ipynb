```python
lst = []
template = """Understand the question and answer accordingly
Human: {human_input}
Assistant:"""

prompt = PromptTemplate(input_variables=["human_input"], template=template)
```

```python
element = LLMChain(
    llm=OpenAI(model_name="gpt-3.5-turbo", temperature=0.2, max_tokens=2000),
    prompt=prompt,
    verbose=True,
    memory=ConversationBufferWindowMemory(k=2),
)
```

```python
output2 = element.predict(human_input=text + " Summarize this as in active voice")
lst.append("Summary: " + output2)

output2 = element.predict(human_input=text + " Give me the name of the person in the text. Output only the name")
lst.append("Name of the Person: " + output2)

output = element.predict(human_input=text + " What problem the customer faced as mentioned in the text?")
lst.append("Problem: " + output)

output6 = element.predict(human_input=text + " What is the policy number or compliant number mentioned in the text?")
lst.append("Policy number: " + output6)

output3 = element.predict(
    human_input=text + " Has the person contacted the company before for the same problem as per the text? Answer in yes or no")

if 'yes' in output3.lower():
    output4 = element.predict(
        human_input=text + " How the person contacted the company before for the same problem as per the text?")
    lst.append("Has reported before: " + output4)

s = ''
for i in lst:
    s += "- " + i + "\n"

print(s)
```
